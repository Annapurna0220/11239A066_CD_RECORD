a) School Database

```python
# Symbol table using dictionary
symbol_table = {}

n = int(input("Enter number of students: "))

for i in range(n):
    name = input("Enter student name: ")
    roll = input("Enter roll number: ")
    symbol_table[name] = roll

print("Symbol Table:")
for key, value in symbol_table.items():
    print(key, ":", value)1
```

b) Bakery Inventory

```python
symbol_table = {}

n = int(input("Enter number of items: "))

for i in range(n):
    item = input("Enter item name: ")

    try:
        price = int(input("Enter price: "))
        symbol_table[item] = price
    except:
        print("Invalid price! Enter number only.")
        break

print("Symbol Table:")
for key, value in symbol_table.items():
    print(key, ":", value)
```
