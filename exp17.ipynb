Check if expression is correctly parenthesized

```python
def is_balanced(expr):
    stack = []
    for ch in expr:
        if ch == '(':
            stack.append(ch)
        elif ch == ')':
            if not stack:
                return False
            stack.pop()
    return len(stack) == 0

print(is_balanced("(a+b))"))          # False
print(is_balanced("((a+b)+(c+d))"))  # True
```
