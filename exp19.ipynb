Count spaces and number of lines in a file

```python
def count_spaces_lines(text):
    lines = text.count('\n') + 1
    spaces = text.count(' ')
    return lines, spaces

# Give text directly
text = """Hello world
This is a test
Count spaces here"""

lines, spaces = count_spaces_lines(text)

print("Number of lines:", lines)
print("Number of spaces:", spaces)
```
