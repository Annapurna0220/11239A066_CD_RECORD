a) Banking Transaction Log Example

```python
def is_balanced(expr):
    stack = []
    pairs = {')': '(', '}': '{', ']': '['}

    for char in expr:
        if char in "({[":
            stack.append(char)
        elif char in ")}]":
            if not stack or stack[-1] != pairs[char]:
                return False
            stack.pop()

    return len(stack) == 0


def banking_log_check():
    expr = input("Enter Banking Transaction Expression: ")

    if is_balanced(expr):
        print("Valid Banking Transaction Log (Parentheses Balanced)")
    else:
        print("Invalid Banking Transaction Log (Parentheses Not Balanced)")


# Run
banking_log_check()
```

b) School Mark Sheet Formula

```python
def marksheet_check():
    expr = input("Enter Mark Sheet Formula: ")

    if is_balanced(expr):
        print("Valid Mark Sheet Formula (Parentheses Balanced)")
    else:
        print("Invalid Mark Sheet Formula (Parentheses Not Balanced)")


# Run
marksheet_check()
```
