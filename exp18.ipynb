Check if string is a keyword

```python
def is_keyword(word, keywords):
    return word.upper() in keywords

keywords = {"ORDER", "CANCEL", "TRACK"}
print(is_keyword("TRACK", keywords))   # True
print(is_keyword("pay", keywords))     # False
```
