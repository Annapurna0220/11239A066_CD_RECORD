Check whether a given identifier is valid or not

```python
import re

def is_valid_identifier(name):
    pattern = r'^[A-Za-z][A-Za-z0-9_]*$'
    return bool(re.fullmatch(pattern, name))

# Test cases
print(is_valid_identifier("Cake_01"))   # True
print(is_valid_identifier("_Cake"))     # False
print(is_valid_identifier("Cake01"))    # True
print(is_valid_identifier("1Cake"))     # False
```
